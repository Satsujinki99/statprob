```python
import matplotlib.pyplot as plt

# Data jumlah gol
gol = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]

# Mencari nilai tertinggi, terendah, dan rentang
tertinggi = max(gol)
terendah = min(gol)
rentang = tertinggi - terendah

print(f'Nilai Tertinggi: {tertinggi}')
print(f'Nilai Terendah: {terendah}')
print(f'Rentang: {rentang}')

# Membuat plot
plt.figure(figsize=(10,5))
plt.bar(range(len(gol)), gol)
plt.title('Jumlah Gol Pemain Sepak Bola')
plt.xlabel('Pemain')
plt.ylabel('Jumlah Gol')
plt.grid(True)

# Menambahkan legenda untuk rentang
plt.legend([f'Rentang: {rentang}'])

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data gol
gol = np.array([19, 10, 10, 12, 13, 15, 14, 13, 9, 17])

# Menghitung rata-rata
rata_rata = np.mean(gol)
print(f"Rata-rata: {rata_rata}")

# Menghitung kuadrat selisih setiap nilai data dengan rata-rata
kuadrat_selisih = (gol - rata_rata) ** 2
print(f"Kuadrat selisih: {kuadrat_selisih}")

# Menghitung variansi
variansi = np.var(gol)
print(f"Variansi: {variansi}")

# Menghitung standar deviasi
std_dev = np.std(gol)
print(f"Standar Deviasi: {std_dev}")

# Membuat grafik balok
plt.bar(range(len(gol)), gol)
plt.axhline(y=rata_rata, color='r', linestyle='-')
plt.axhline(y=rata_rata + std_dev, color='g', linestyle='--')
plt.axhline(y=rata_rata - std_dev, color='g', linestyle='--')

# Menambahkan legenda
plt.legend([f'Rata-rata: {rata_rata}'])
plt.legend([f'Standar deviaasi: {std_dev}'])
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data tinggi pemain
tinggi = np.array([180, 185, 175, 190, 170, 195, 165])

# Menghitung nilai tertinggi dan terendah
nilai_tertinggi = np.max(tinggi)
nilai_terendah = np.min(tinggi)
print(f"Nilai Tertinggi: {nilai_tertinggi} cm")
print(f"Nilai Terendah: {nilai_terendah} cm")

# Menghitung rentang
rentang = nilai_tertinggi - nilai_terendah
print(f"Rentang: {rentang} cm")

# Membuat grafik balok
plt.bar(range(len(tinggi)), tinggi)
plt.axhline(y=nilai_terendah, color='r', linestyle='-')
plt.axhline(y=nilai_tertinggi, color='g', linestyle='-')

# Menambahkan legenda
plt.legend(["Nilai Terendah", "Nilai Tertinggi"])
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data tinggi pemain
tinggi = np.array([180, 185, 175, 190, 170, 195, 165])

# Menghitung rata-rata
rata_rata = np.mean(tinggi)
print(f"Rata-rata: {rata_rata} cm")

# Menghitung kuadrat selisih setiap nilai data dengan rata-rata
kuadrat_selisih = (tinggi - rata_rata) ** 2
print(f"Kuadrat selisih: {kuadrat_selisih}")

# Menghitung variansi
variansi = np.var(tinggi)
print(f"Variansi: {variansi}")

# Membuat grafik balok
plt.bar(range(len(tinggi)), tinggi)
plt.axhline(y=rata_rata, color='r', linestyle='-')
plt.axhline(y=rata_rata + np.sqrt(variansi), color='g', linestyle='--')
plt.axhline(y=rata_rata - np.sqrt(variansi), color='g', linestyle='--')

# Menambahkan legenda
plt.legend(["Rata-rata", "Variansi"])
plt.grid(True)
plt.show()
```
