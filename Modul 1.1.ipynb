```python
import matplotlib.pyplot as plt
import numpy as np

# Data tinggi badan siswa
data = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]

# Menghitung mean
mean = np.mean(data)

# Menampilkan data
print(f"Data tinggi badan siswa: {data}")

# Menghitung jumlah semua nilai
sum_data = np.sum(data)
print(f"Jumlah semua nilai: {sum_data} cm")

# Menghitung jumlah total data
count_data = len(data)
print(f"Jumlah total data: {count_data}")

# Mencetak mean
print(f"Mean (Rata-rata) tinggi badan siswa: {mean} cm")

# Membuat plot
plt.figure(figsize=(10,5))
plt.hist(data, bins=20, color='blue', edgecolor='black')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean} cm')
plt.title('Distribusi Tinggi Badan Siswa Kelas A')
plt.xlabel('Tinggi Badan (cm)')
plt.ylabel('Frekuensi')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data tinggi badan siswa
data = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]

# Menghitung median
median = np.median(data)

# Menampilkan data
print(f"Data: {data}")

# Menampilkan data berurut
data.sort()
print(f"Data berurut: {data}")

# Menghitung jumlah total data
count_data = len(data)
print(f"Jumlah total data: {count_data}")

# Menampilkan nilai tengah (median)
print(f"Nilai tengah (Median): {median} cm")


# Membuat plot
plt.figure(figsize=(10,5))
plt.hist(data, bins=20, color='blue', edgecolor='black')
plt.axvline(median, color='red', linestyle='dashed', linewidth=2, label=f'Median: {median} cm')
plt.title('Distribusi Tinggi Badan Siswa Kelas A')
plt.xlabel('Tinggi Badan (cm)')
plt.ylabel('Frekuensi')
plt.legend()
plt.grid(True)
plt.show()
```
